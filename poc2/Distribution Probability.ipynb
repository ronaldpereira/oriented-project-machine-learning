```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1212)
```

The probability density for the Gaussian distribution is ![density_formula](assets/density_formula.png)

```python
class gaussianDist:
    def __init__(self, name, mu, sigma, size):
        self.name = name
        self.mu = mu
        self.sigma = sigma
        self.size = size
        self.points = np.random.normal(self.mu, self.sigma, self.size)
        
    def probability(self, x):
        return 1/(self.sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - self.mu)**2 / (2 * self.sigma**2))
        
    def plot(self):
        count, bins, _ = plt.hist(self.points, 30, density=True)
        plt.plot(bins, self.probability(bins), linewidth=2, color='r', label="p(x)")
        plt.legend(loc='best')
        plt.show()
```
